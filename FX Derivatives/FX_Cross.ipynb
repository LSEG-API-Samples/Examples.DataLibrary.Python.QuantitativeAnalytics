## Some Imports to start with

```python
import refinitiv.data as rd
import refinitiv.data.content.ipa.financial_contracts as rdf
from refinitiv.data.content.ipa.financial_contracts import cross
```

## Open the data session

The open_session() function creates and open sessions based on the information contained in the refinitiv-data.config.json configuration file. Please edit this file to set the session type and other parameters required for the session you want to open.

```python
rd.open_session()
```

## FX Cross

### Single

```python
response = cross.Definition(

    instrument_tag="1Y-CHFJPY",
    fx_cross_type=cross.FxCrossType.FX_SWAP,
    fx_cross_code="CHFJPY",

    legs=[
        cross.LegDefinition(
            deal_ccy_buy_sell="buy",
            fx_leg_type="SwapNear",
            deal_amount=1000000,
            contra_amount=897008.3,
            tenor="1M"
        ),
        cross.LegDefinition(
            deal_ccy_buy_sell="sell",
            fx_leg_type="SwapFar",
            deal_amount=1000000,
            contra_amount=900000,
            tenor="1Y"
        ),
    ],

).get_data()

response.data.df
```

### Multiple

```python
response = rdf.Definitions(

    universe = [
        cross.Definition(
            instrument_tag = "1Y-CHFJPY",
            fx_cross_type = cross.FxCrossType.FX_SWAP,
            fx_cross_code = "CHFJPY",

            legs = [
                cross.LegDefinition(
                    deal_ccy_buy_sell = "buy",
                    fx_leg_type = "SwapNear",
                    deal_amount = 1000000,
                    contra_amount = 897008.3,
                    tenor = "1M"
                ),
                cross.LegDefinition(
                    deal_ccy_buy_sell = "sell",
                    fx_leg_type = "SwapFar",
                    deal_amount = 1000000,
                    contra_amount = 900000,
                    tenor = "1Y"
                ),
            ])
    ]
).get_data()

response.data.df
```

### Fields selection

```python
response = cross.Definition(

    instrument_tag = "1Y-CHFJPY",
    fx_cross_type = cross.FxCrossType.FX_SWAP,
    fx_cross_code = "CHFJPY",
    legs = [
        cross.LegDefinition(
            deal_ccy_buy_sell = "buy",
            fx_leg_type = "SwapNear",
            deal_amount = 1000000,
            contra_amount = 897008.3,
            tenor = "1M"
        ),
        cross.LegDefinition(
            deal_ccy_buy_sell = "sell",
            fx_leg_type = "SwapFar",
            deal_amount = 1000000,
            contra_amount = 900000,
            tenor = "1Y"
        ),
    ],
    fields = [
        "InstrumentTag",
        "ValuationDate",
        "InstrumentDescription",
        "EndDate",
        "FxSwapsCcy1",
        "FxSwapsCcy2",
        "FxSwapsCcy1Ccy2",
        "FxOutrightCcy1Ccy2",
        "ErrorCode",
        "ErrorMessage"
    ]


).get_data()

response.data.df
```

### Global calculation parameters

```python
response = cross.Definition(
    
    instrument_tag = "1Y-CHFJPY",
    fx_cross_type = cross.FxCrossType.FX_SWAP,
    fx_cross_code = "CHFJPY",
    legs = [
        cross.LegDefinition(
            deal_ccy_buy_sell = "buy",
            fx_leg_type = "SwapNear",
            deal_amount = 1000000,
            contra_amount = 897008.3,
            tenor = "1M"
        ),
        cross.LegDefinition(
            deal_ccy_buy_sell = "sell",
            fx_leg_type = "SwapFar",
            deal_amount = 1000000,
            contra_amount = 900000,
            tenor = "1Y"
        ),
    ],
    pricing_parameters = cross.PricingParameters(
        valuation_date = "2018-02-17T00:00:00Z",
        price_side = cross.PriceSide.ASK
    )

).get_data()

response.data.df
```

### Global & individual calculation parameters

```python
response =  rdf.Definitions(
    universe = [cross.Definition(
        instrument_tag = "1Y-CHFJPY",
        fx_cross_type = cross.FxCrossType.FX_SWAP,
        fx_cross_code = "CHFJPY",
        legs = [
            cross.LegDefinition(
                deal_ccy_buy_sell = "buy",
                fx_leg_type = "SwapNear",
                deal_amount = 1000000,
                contra_amount = 897008.3,
                tenor = "1M"
            ),
            cross.LegDefinition(
                deal_ccy_buy_sell = "sell",
                fx_leg_type = "SwapFar",
                deal_amount = 1000000,
                contra_amount = 900000,
                tenor = "1Y"
            ),
        ],
        pricing_parameters = cross.PricingParameters(
            valuation_date = "2018-02-17T00:00:00Z",
            price_side = cross.PriceSide.ASK
        )

    )],
    pricing_parameters = cross.PricingParameters(
        valuation_date = "2018-02-17T00:00:00Z",
        price_side = cross.PriceSide.ASK
    )).get_data()

response.data.df
```
